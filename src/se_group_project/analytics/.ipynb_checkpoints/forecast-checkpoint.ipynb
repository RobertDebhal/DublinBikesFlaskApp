```python
import datetime
import math
intercept = 2.9882352941176498
coef = [-0.4705882352941184, 0.0, 0.0, 0.0] 
coef_wind = coef[0]
coef_temp=  coef[1]
coef_rain_t=  coef[2]
coef_rain_f= coef[3]

new_future=[]
day = 3 #input data
hour = 8

index=0
for i in range(len(future['list'])):
    date_time_object = future['list'][i]['dt_txt']
    date_time_list=date_time_object.split(' ')
    #print(date_time_list) 
    date=date_time_list[0].split()
    date_seperated = date[0].split('-')
    year = date_seperated[0]
    month = date_seperated[1]
    day1 = date_seperated[2]
    hour_list= date_time_list[1].split(':')
    hour_of_day= hour_list[0]
    #print(year,month,day,hour_of_day)
    myObj = datetime.datetime(int(year),int(month),int(day1),int(hour_of_day))
    
    print(myObj.weekday(),day)
    if myObj.weekday()==day:
        new_future.append(future['list'][i])
        new_future[index]['dt_txt_new']=myObj
        index+=1
    temp = future['list'][i]['main']['temp'] -273.15
    rain_f_boolean = future['list'][i]['weather'][0]['description']!='moderate rain' # True when no rain
    rain_f = rain_f_boolean*1
    rain_t_boolean = future['list'][i]['weather'][0]['description']=='moderate rain' # True when no rain
    rain_t = rain_t_boolean*1
    wind = future['list'][i]['wind']['speed']
    pred = intercept + (coef_wind*wind) + (temp* coef_temp) + (rain_f*coef_rain_f) + (rain_t*coef_rain_t )
    #print(coef_wind, coef_temp,coef_rain_t, coef_rain_f)
    actual_pred = 0 
    if pred <0:  # number of bikes at a station cant be negative 
        actual_pred=0
    else:
        actual_pred =math.ceil(pred)

def get_closest_hour(num):       
    nums = [[0,3],[3,6],[6,9],[9,12],[12,15],[15,18],[18,21]]
    found=False
    for elem in nums:
        if num > elem[0]and num<elem[1]:
            if (num - elem[0])<=1.5:
                num=elem[0]
            else:
                num=elem[1]
            found=True
    if found:
        return num
    else:
        if num-21<=1.5:
            num=21
        else:
            num=0
        return num
    
adjusted_hour=get_closest_hour(hour)     

index_of_hour=0
for i in range(len(new_future)):
    if new_future[i]['dt_txt_new'].hour==adjusted_hour:
        index_of_hour=i
        

    
    
        
print(new_future[index_of_hour])
#print('our dictionary is', new_future[0])
```
